```
def clean_data(simbolData):
    simbolData = re.sub(r'[?|$|.|!|\n|;|-|,|:|\r|\d+]',r'',simbolData)
    simbolData = html.parser.unescape(simbolData)
    return simbolData
```

```
length = review.size
for i in range(0,length):
    review[i] = clean_data(review.item(i))
```

```
review
```

```
review = [item.lower() for item in review]
```

```
review
```

```
sentiment_analyzer = SentimentIntensityAnalyzer()
positive_rating = []
negitive_rating = []
n_records=len(review)
for i in range(0,length):
     score = sentiment_analyzer.polarity_scores(review[i])
     positive_rating.append(score['pos']*100)
     negitive_rating.append(score['neg']*100)
```

```
positive_rating=np.array(positive_rating)
negitive_rating=np.array(negitive_rating)
```

```
output = pd.DataFrame({
    'Positive':positive_rating[0:length],
    'Negative':negitive_rating[0:length]
})
```

```
def compute_rating(pos,neg):
    rating = 0
    if pos > 30:
        rating =5
    elif pos >20:
        rating =4
    elif pos >10:
        rating =3
    elif pos >0:
        rating =2
        
    penalty=0
    if neg>5 and neg<=20:
        penalty=1
    elif neg>20 and neg<=100:
        penalty=2
    
    rating= rating-penalty
    if rating<0:
        rating=0
    return rating
```

```
def predict_rating(df):
    rating_array = []
    for index, row in df.iterrows():
        rating=compute_rating(row['Positive'],row['Negative'])
        rating_array.append(rating)
    return rating_array
```

```
rating_array =predict_rating(hasil)
```
